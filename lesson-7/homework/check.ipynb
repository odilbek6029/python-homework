```python
class Vector:
    def __init__(self,*arg):
        self.x=arg
           
        
    def __str__(self):
        return f"Vector{self.x}"

    def __len__(self):
        sum=0
        for i in range(len(self.x)):
           sum+=1 
        return (sum)
    
    def __add__(self, other):
        addi=[self.x[i]+other.x[i] for i in range(len(self.x))]
        return Vector(*addi)
    
    def __sub__(self, other):
        if len(self.x)==len(other.x):
         subt=[self.x[i]-other.x[i] for i in range(len(self.x))]
         return Vector(*subt)
        else:
         raise ValueError("Two vectors have different dimension")
         
    def __mul__(self, other):
       dot=0
       if len(self.x)!=len(other.x):
          raise ValueError("Two vectors have different dimension")
       for i in range(len(self.x)):
          dot+=self.x[i]*other.x[i]
       return dot
       
    
    def __rmul__(self, y):
            mult=[self.x[i]*y for i in range(len(self.x))]
            return Vector(*mult)
    
    def magnitude(self):
        mag=0
        for i in range(len(self.x)):
            mag+=self.x[i]**2
        return mag**(1/2)
    
    def normalize(self):
        norm=[]
        for i in range(len(self.x)):
            norm.append(self.x[i]/self.magnitude())
        return Vector(*norm)


a=Vector(3,4)
b=Vector(1,2,4)
print(len(a))
print(a.normalize())
```

```python
class Tasks:
    def __init__(self, task_id, title, description, due_date, status):
        self.task_id=task_id
        self.title=title
        self.description=description
        self.due_date=due_date
        self.status=status

    def __str__(self):
         return f"{self.task_id}, {self.title}, {self.description}, {self.due_date}, {self.status}"
    
   
class TaskManager:
    def __init__(self):
        self.tasks=[]
    def __iter__(self):
        return iter(self.tasks)

    def add_task(self, task_id, title, description, due_date, status):
        self.tasks.append(Tasks(task_id, title, description, due_date, status))
    def view_tasks(self):
     if not self.tasks:
        print("There is nothing")
     else: 
        print(f"Employee records:")
        print("ID, title, description, due date, status")
        for task in self.tasks:  
          print(f"{task}")
    
    def update(self, task_id):
     for task in self.tasks: 
        if task.task_id == task_id:  # to check task_id
            task.task_id = input("Enter updated ID: ")
            task.title = input("Enter a title: ")
            task.description = input("Enter a description: ")
            task.due_date = input("Enter a due date: ")
            task.status = input("Enter a status: ")
            print(f"Updated successfully!")

    def delete(self,task_id):
        self.tasks=[tasks for tasks in self.tasks if tasks.task_id!=task_id] 
        print("This data has just deleted! ")       

    def filter_task(self):
        pend=[tasks for tasks in self.tasks if tasks.status=="pending"]
        comp=[tasks for tasks in self.tasks if tasks.status=="completed"]     
        combined=pend+comp
        print("ID, title, description, due date, status")
        for comb in combined:
            print(f"{comb}")
    
    def save_tasks(self):
        with open("file.csv", "a")as f:
           for task in self.tasks: 
            f.write(f"{task}\n")

    def load_tasks(self):
        with open("file.csv", "r") as f:
            txt=f.readlines()
        for lines in txt:
            sample=lines.strip().split(",")
            self.tasks.append(Tasks(*sample))   

       

options="""1. Add a new task
2. View all tasks
3. Update a task
4. Delete a task
5. Filter tasks by status
6. Save tasks
7. Load tasks
8. Exit"""
print(options)     
manager=TaskManager()
while True:
            
 opt=int(input("Choose an option: "))
 if opt==1:
        id=input("Enter taskID: ")
        tit=input("Enter title: ")
        desc=input("Enter description: ")
        due=input("Enter due date: ")
        stat=input("Enter a status: ")
        manager.add_task(id, tit, desc, due,stat)
        print("Task added successfully!")

 if opt==2:
        manager.view_tasks()

 if opt==3:
     id=input("Enter ID to update: ")
     manager.update(id)

 if opt==4:
     id=input("Enter ID to delete: ")
     manager.delete(id)

 if opt==5:    
     manager.filter_task()

 if opt==6:
     manager.save_tasks()

 if opt==7:
     manager.load_tasks()
 
 if opt==8:
     print("Goodbye!")
     break
```
