```python
def check(func):
    def wrap(ar,key):
         try:
              func(ar,key)
         except ZeroDivisionError:
              print("Denominator can't be zero")
    return wrap


@check

def division(a, b):
     print(a/b)

division(3, 0)  
```

```python
import string

with open("sample.txt","r") as f:
    txt=f.readlines()
all_words=[]
for line in txt:
    test=line.replace("\n", " ").translate(str.maketrans('', '', string.punctuation)).lower() #to remove punctuations
    words=test.split(" ")
    all_words.extend(words)

all_words=[words for words in all_words if words!=""] #to remove ""
tot_number=len(all_words) #total number of the words
freq_word=[]
num=[]

for word in all_words: #to find distinguished words
    if word not in freq_word:
        freq_word.append(word)
    else:
        pass
for i in range(tot_number):
    if all_words[i] in freq_word:
        num.append((all_words.count(all_words[i]),f"{all_words[i]}"))
        num.sort()
        freq_word.remove(all_words[i])
    
freq_word=[num[-1],num[-2],num[-3],num[-4],num[-5]]# find 5 most common words and its number

component=f"Total words: {tot_number}\nTop 5 most common words:\n{freq_word[0][1]}: {freq_word[0][0]} times\n{freq_word[1][1]}: {freq_word[1][0]} times\n{freq_word[2][1]}: {freq_word[2][0]} times\n{freq_word[3][1]}: {freq_word[3][0]} times\n{freq_word[4][1]}: {freq_word[4][0]} times"
with open("word_count_report.txt", "w") as f:
    f.write(component)
```
